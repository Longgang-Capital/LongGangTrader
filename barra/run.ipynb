```python
import pandas as pd

def filter_df_by_date(df: pd.DataFrame, min_date, date_col: str = "date") -> pd.DataFrame:
    d = pd.to_datetime(min_date)
    c = df[date_col]
    if not pd.api.types.is_datetime64_any_dtype(c):
        c = pd.to_datetime(c, errors="coerce")
    res = df.loc[c > d].copy()
    return res.reset_index(drop=True)
```

```python
style = filter_df_by_date(style_factors, pd.to_datetime('20231001'))
industry = filter_df_by_date(industry_factors, pd.to_datetime('20231001'))
mcaps = filter_df_by_date(mcaps, pd.to_datetime('20231001'))
rets = filter_df_by_date(rets, pd.to_datetime('20231001'))
portfolio = filter_df_by_date(portfolio, pd.to_datetime('20231001'))
```
